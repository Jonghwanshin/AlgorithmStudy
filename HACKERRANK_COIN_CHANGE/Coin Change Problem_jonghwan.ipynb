```python
#!/bin/python3

import sys

cache = [-1 for x in range(250)]

def getWays(n, c):
    # Complete this function
    #print("n :",n)
    if(n <= 0) or (c[0] > n):
        return 0
    if(cache[n] != -1):
        return cache[n]
    if(n in c):
        if(cache[n] == -1):
            cache[n] = 1
        else:
            cache[n] = cache[n] + 1
    idx = 0
    while((n - c[idx]) > 0):
        print(idx)
        cache[n] = getWays(n-c[idx],c) + getWays(c[idx],c)
        idx = idx +1
        #return cache[n]
    return cache[n]

n, m = input().strip().split(' ')
n, m = [int(n), int(m)]
c = list(map(int, input().strip().split(' ')))
c.sort()
print(c)
# Print the number of ways of making change for 'n' units using coins having the values given by 'c'
ways = getWays(n, c)
print(cache[1:250])
print(ways)
```
